# **Chain**
**Chains enable connecting prompts with LLMs to create structured workflows.**

```python
from langchain_core.prompts import PromptTemplate

# Define prompt
template = PromptTemplate.from_template(
    "Explain {topic} in simple terms"
)

# Simulate chaining logic
def simple_chain(input_data):
    prompt = template.format(**input_data)
    return f"[LLM Response Simulation]: {prompt}"

print(simple_chain({"topic": "Deep Learning"}))
```

# **Memory**

**Memory helps in storing past interactions and enables context-aware responses.**

```python
# Simple memory simulation
memory = []

def save_memory(user_input, response):
    memory.append({"input": user_input, "output": response})

def load_memory():
    return memory

# Example usage
save_memory("Hi", "Hello")
print(load_memory())
```
