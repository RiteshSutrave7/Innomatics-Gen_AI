## **Task 1: Using PromptTemplate instead of f-string**

```python
template1 = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

print(template1.format(topic="Artificial Intelligence"))
```

# **Task 2: Multiple inputs**

```python
template2 = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

print(template2.format(topic="AI", audience="beginners", tone="friendly"))
print(template2.format(topic="Python", audience="kids", tone="fun"))
print(template2.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

# **Task 3: Different styles**

```python
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

# **Task 4: Chat roles**

```python
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {role}"),
    ("user", "Explain {topic}")
])

messages = chat_template.format_messages(
    role="teacher",
    topic="Neural Networks"
)

print(messages)
```

# **Task 5: Validation function**

```python
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience! Default set to beginner.")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone! Default set to formal.")
        tone = "formal"

    return audience, tone

# Test
print(validate_inputs("student", "cool"))
```

# **Task 6 – Prompt Generator App**

```python
# Task 6: Dynamic prompt generator

def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    templates = {
        "teaching": "Explain {topic} clearly for {audience} in a {tone} tone",
        "interview": "Ask 3 questions about {topic} for {audience}",
        "storytelling": "Explain {topic} as a story in a {tone} tone"
    }

    prompt_template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=templates[style]
    )

    return prompt_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

# Test
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

# **Task 7: Reusability**


```python
reuse_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

topics = ["AI", "Python", "Blockchain", "IoT", "Cloud Computing"]

for t in topics:
    print(reuse_template.format(topic=t))
```
